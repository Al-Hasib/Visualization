```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.arange(1,10)
y = [i**2 for i in range(1,10)]
```

```python
# Scatter Plot
plt.scatter(x,y,color='g',marker='*')
plt.xlabel("x")
plt.ylabel("y")
plt.title("X Vs Y")
```

```python
# Plot Graph
plt.plot(x,y,marker='+',color='red',linestyle='dashed',markersize=10)
plt.xlabel('X_axis')
plt.ylabel('y_Axis')
plt.title("X Vs Y")
```

```python
# create Subplot
plt.subplot(2,2,1)
plt.plot(x,y,color='blue')
plt.subplot(2,2,2)
plt.scatter(x,y,color='blue')
plt.subplot(2,2,3)
plt.plot(x,y,color='red')
plt.subplot(2,2,4)
plt.scatter(x,y,color='r')
```

```python
# Bar plot
plt.bar(x,y,color='green')
plt.xlabel("X Axis")
plt.ylabel("y Axis")
plt.title("X Vs Y")
plt.show()
```

```python
# Histogram
a = [2,5,34,56,4,54,20,24,34,67,89,34,45,2,13,46,87]
plt.hist(a,bins=20)
plt.title('count of a')
plt.xlabel(a)
```

```python
# Boxplot
a = [2,5,34,56,4,54,20,24,34,67,89,34,45,2,13,46,87]
plt.boxplot(a)
plt.title('Boxplot of A')
plt.show()
```

```python
# Pie Chart
labels = ['python','c++','ruby','java']
sizes = [213,132,100,123]
colors = ['gold','blue','green','orange']
explode = (0,0,0.1,0)
plt.pie(sizes,labels=labels,explode=explode,colors=colors,shadow=True,autopct='%1.2f%%')
plt.axis('equal')
plt.show()
```
