```python
import seaborn as sns
df = sns.load_dataset('tips')
# for continuous data,we should use jointplot,pairplot,distplot
# for categorical data,we should use boxplot,violinplot,countplot,barplot
```

```python
df.head(3)
```

```python
df.corr(method='spearman')
```

```python
sns.heatmap(df.corr(method='spearman'))
```

```python
# univariate analysis
# jointplot
sns.jointplot(x='tip',y='total_bill',data=df,kind='scatter')
```

```python
# paiplot
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')
```

```python
# distplot works as histogram
sns.distplot(df['tip'])
```
